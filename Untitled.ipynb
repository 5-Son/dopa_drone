```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

RESULT_DIR = "results"

def load_all_results(result_dir=RESULT_DIR):
    rows = []
    raw_data = []

    for fname in sorted(os.listdir(result_dir)):
        if not fname.endswith(".json"):
            continue
        path = os.path.join(result_dir, fname)
        with open(path, "r") as f:
            data = json.load(f)
        data["_filename"] = fname
        raw_data.append(data)

        pareto = data.get("final_pareto", [])
        row = {
            "filename": fname,
            "scenario": data.get("scenario", "NA"),
            "scenario_name": data.get("scenario_name", ""),
            "seed": data.get("seed", None),
            "exec_time": data.get("execution_time", np.nan),
            "pareto_size": len(pareto),
            "entropy_last": data.get("entropy", np.nan),
        }
        rows.append(row)

    df_runs = pd.DataFrame(rows)
    return df_runs, raw_data

df_runs, raw_results = load_all_results()
df_runs
```

```python
# 시나리오별 실행시간, Pareto 크기 통계
df_summary = (
    df_runs
    .groupby("scenario")
    .agg(
        num_runs=("filename", "count"),
        exec_mean=("exec_time", "mean"),
        exec_std=("exec_time", "std"),
        pareto_mean=("pareto_size", "mean"),
        pareto_std=("pareto_size", "std"),
        entropy_mean=("entropy_last", "mean"),
        entropy_std=("entropy_last", "std"),
    )
    .reset_index()
)

df_summary
```

```python
rows_pf = []

for data in raw_results:
    scenario = data.get("scenario", "NA")
    scenario_name = data.get("scenario_name", "")
    seed = data.get("seed", None)
    fname = data.get("_filename", "")
    pareto = data.get("final_pareto", [])

    for f in pareto:
        # f = [F1, F2, F3]
        if len(f) < 3:
            continue
        rows_pf.append({
            "scenario": scenario,
            "scenario_name": scenario_name,
            "seed": seed,
            "filename": fname,
            "F1": f[0],
            "F2": f[1],
            "F3": f[2],
        })

df_pf = pd.DataFrame(rows_pf)
df_pf.head()
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
df_runs.boxplot(column="exec_time", by="scenario")
plt.title("Execution Time by Scenario")
plt.suptitle("")
plt.xlabel("Scenario")
plt.ylabel("Time [s]")

plt.subplot(1,2,2)
df_runs.boxplot(column="pareto_size", by="scenario")
plt.title("Pareto Size by Scenario")
plt.suptitle("")
plt.xlabel("Scenario")
plt.ylabel("Number of Pareto Solutions")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,5))

for scen, group in df_pf.groupby("scenario"):
    plt.scatter(group["F1"], group["F2"], s=10, alpha=0.5, label=scen)

plt.xlabel("F1 (Strike Effectiveness)")
plt.ylabel("F2 (Detection Time)")
plt.title("F1 vs F2 Pareto Front (All Scenarios)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
scenario_to_view = "S4"  # "S1", "S2", "S3"으로 바꿔도 됨

g = df_pf[df_pf["scenario"] == scenario_to_view]

plt.figure(figsize=(6,5))
plt.scatter(g["F1"], g["F2"], s=10, alpha=0.7)
plt.xlabel("F1 (Strike Effectiveness)")
plt.ylabel("F2 (Detection Time)")
plt.title(f"Pareto Front (Scenario {scenario_to_view})")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 예시: S4, seed0 선택
target_scen = "S4"
target_seed = 0

r = None
for data in raw_results:
    if data.get("scenario") == target_scen and data.get("seed") == target_seed:
        r = data
        break

if r is None:
    print("해당 시나리오/seed 조합 결과를 찾을 수 없습니다.")
else:
    w_trace = r.get("wasserstein_trace", [])
    e_trace = r.get("entropy_trace", [])

    plt.figure(figsize=(10,4))
    
    plt.subplot(1,2,1)
    plt.plot(w_trace)
    plt.title(f"Wasserstein Trace (Scenario {target_scen}, seed {target_seed})")
    plt.xlabel("Generation")
    plt.ylabel("W1 distance")
    plt.grid(True)

    plt.subplot(1,2,2)
    plt.plot(e_trace)
    plt.title(f"Entropy Trace (Scenario {target_scen}, seed {target_seed})")
    plt.xlabel("Generation")
    plt.ylabel("Entropy")
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```
